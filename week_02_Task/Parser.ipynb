Napr. Behem celeho minuleho roku jsme zkoumali deformaci prvku v zavislosti na teplote.
Prvky jsou oznaceny symboly A, B, C, D
Kazde mereni system zaznamenaval do souboru (experiment_02_01.csv) ve formatu: Datum; Prvek; Teplota; Deformace
Mereni nebyly serazene v case
Zaznam muze obsahovat chyby:
chybny prvek (viz. radek 5, prvek F)
deformace obsahuje chybnou carku misto tecky (viz. radek 1, 0,92)
teplota muze byt mimo povoleny rozsah (viz. radek 19)

Ukolem bude napsat parser vstupniho souboru csv. 

Parser by mel:
filtrovat hodnoty dle data a dle zdroje
kontrolovat hodnoty na pripustny rozsah teplot a deformaci
eliminovat chyby

Vysledek by se mel ukladat do finalniho souboru dle formatu (viz. vzor). Datum se uklada ve vterinach (viz. https://www.w3schools.com/python/python_datetime.asp)


```python
from datetime import datetime
# Reading data from file to list
name_file_input = 'experiment_02_01.csv'

with open(name_file_input, 'r', encoding="utf-8") as f:
    raw_data = [line.rstrip() for line in f]

raw_data[:10]
```

```python
# Parsing data
parsing_data = []

for line in raw_data:
    raw_data = line.split(';')
    parsing_data.append(raw_data)

parsing_data
parsing_data[:10]
```

```python
   
# user function for cleaning
def cleaning_data(input_value, input_type, min_value=None, max_value=None):

    input_value = input_value.strip()
    # clearing int and float value
    if input_type == int or input_type == float:
        input_value = input_value.replace(',', '.')
        try: 
            input_value = input_type(input_value)
        except:
            return False
        
        if min_value:
            if input_value < min_value:
                return False
        
        if max_value:
            if input_value > max_value:
                return False
                
        
        return input_value
            
    # clearing string value
    if input_type == str:

        type_list = ['A', 'B', 'C', 'D']

        input_value = input_value.upper()
        if input_value not in type_list:
            return False
        
        return input_value
    
    return False
```

Zadat min. a max. hodnoty teploty a deformace:

```python
min_teplota = -15
max_teplota = 50
min_deformace = -1
max_deformace = 2
```

```python
parsing_data[:10]
```

```python
#  cleaning data


number_of_errors = 0
output_list = []
output_structure = ['Datum', 'Prvek', 'Teplota', 'Deformace']
id_number = 0

for item in parsing_data:

    error_in_line = 0
    output_dict_line = {}
    for key, value in enumerate(item):

        if key == 0:
            #  datum
            data_value = datetime.strptime(value, '%Y-%m-%d %H:%M:%S')
            clean_value = int((data_value - datetime(1970, 1, 1)).total_seconds())
        elif key == 1:
            #  prvek
            clean_value = cleaning_data(value, str)
        elif key == 2:
            #  teplota
            clean_value = cleaning_data(value, float, min_teplota, max_teplota)
        elif key == 3:
            #  deformace
            clean_value = cleaning_data(value, float, min_deformace, max_deformace)
        else:
            clean_value = False
            
        if clean_value == False:
            error_in_line = 1
            break

        output_dict_line[output_structure[key]] = clean_value
        print(f'{output_dict_line=}')
    
    if error_in_line == 0:
        id_number += 1
        output_list.append(output_dict_line)
    else:
        number_of_errors += 1
        
print(f'File has {number_of_errors} errors')
print(f'{len(output_list)} rows were converted')
output_list[:10]
```
