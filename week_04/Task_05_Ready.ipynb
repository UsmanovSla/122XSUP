##### Identify and handle missing values

```python
# identify these missing values: create new dataframe missing_data
missing_value = df.isnull()
missing_value
```

##### Count missing values in each column

```python
# write a function for counting missing value in dataframe

def count_missing(missing_data: pd.DataFrame) -> int:
    for column in missing_data.columns.values.tolist():
        print(column)
        print(missing_data[column].value_counts())
        print("")
```

```python
# call function and print report about missing value in dataframe
count_missing(missing_value)
```

```python
print(f'Missing value: {df.isnull().sum()}')
```

##### Correct data format

* check all data is in the correct format (int, float, text or other)
* change **float** to **int** in column 'Water'

```python
# check data in dataframe
df.dtypes
```

```python
# change data type
df[['Water']] = df[['Water']].astype('int64')
df
```

##### Calculation and operetion with dataframe

* create new column in dataframe 'Hi Strength'
* fill new collum with:
    * 1 if strength >= 30 MPa
    * 0 if strength < 30 MPa

```python
df.loc[df['Strength'] >= 30, 'Hi strength'] = 1
df.loc[df['Strength'] < 30, 'Hi strength'] = 0
df.head()
```

##### Save data as new dataframe to file: data/concrete_data_clean.csv

```python
# Save dataframe to file
df.to_csv("data/concrete_data_clean.csv", index=False)
```
